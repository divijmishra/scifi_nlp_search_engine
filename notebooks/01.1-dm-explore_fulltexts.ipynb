```python
import os
import re
```

```python
# load books fulltext
load_path = "../data/raw/SFGram-dataset-master/book-contents/"

fulltexts = {}

for f in os.listdir(load_path):
    with open(load_path + f) as file:
        fulltext = file.read()

    id = f[5:9]
    fulltexts[id] = fulltext
```
